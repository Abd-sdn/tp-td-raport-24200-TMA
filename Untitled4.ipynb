```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft
from scipy.io import wavfile
from PIL import Image
import os
```

```python
A = 1
f0 = 10
phi = 0
fs = 100
T = 1

t = np.arange(0, T, 1/fs)
x = A * np.sin(2 * np.pi * f0 * t + phi)

plt.figure(figsize=(10,4))
plt.plot(t, x)
plt.title("TP1 - Signal sinusoïdal")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
print("Analyse :")
print("- Le signal est sinusoïdal de fréquence 10 Hz.")
print("- La fréquence d’échantillonnage est 100 Hz.")
print("- Comme fs/f0 = 10, on a 10 échantillons par période.")
```

```python
bruit = np.random.normal(0, 0.3, len(x))
y = x + bruit

plt.figure(figsize=(10,4))
plt.plot(t, x, label="Signal pur")
plt.plot(t, y, label="Signal bruité", alpha=0.8)
plt.title("TP1 - Signal pur et signal bruité")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
print("Analyse :")
print("- Le bruit blanc gaussien perturbe le signal sinusoïdal.")
print("- Le signal bruité conserve la structure globale du signal pur.")
print("- Les fluctuations aléatoires rendent l’observation plus difficile.")
```

```python
N = 100
rect = np.zeros(N)
rect[20:40] = 1

plt.figure(figsize=(10,4))
plt.plot(rect)
plt.title("TP1 - Signal porte")
plt.xlabel("Indice n")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
conv_rect = np.convolve(rect, rect)

plt.figure(figsize=(10,4))
plt.plot(conv_rect)
plt.title("TP1 - Convolution du signal porte avec lui-même")
plt.xlabel("Indice n")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
print("Analyse :")
print("- La convolution de deux signaux portes donne un signal triangulaire.")
print("- Cela s’explique par le recouvrement progressif puis décroissant des deux portes.")
```

```python
fs_audio = 44100
T_audio = 2
t_audio = np.arange(0, T_audio, 1/fs_audio)

f1 = 440
f2 = 880

signal_audio = np.sin(2*np.pi*f1*t_audio) + 0.5*np.sin(2*np.pi*f2*t_audio)

plt.figure(figsize=(10,4))
plt.plot(t_audio[:1000], signal_audio[:1000])
plt.title("TP2 - Signal audio temporel (zoom)")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
N_audio = len(signal_audio)
X_audio = fft(signal_audio)
freq_audio = fftfreq(N_audio, 1/fs_audio)

mask = freq_audio >= 0
freq_pos = freq_audio[mask]
amp_pos = np.abs(X_audio[mask])

plt.figure(figsize=(12,4))
plt.plot(freq_pos, amp_pos)
plt.title("TP2 - Spectre du signal audio")
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.xlim(0, 2000)
plt.grid()
plt.show()
```

```python
print("Analyse :")
print("- Le spectre présente deux pics principaux autour de 440 Hz et 880 Hz.")
print("- Cela correspond exactement aux deux fréquences introduites dans le signal.")
```

```python
f_bruit = 5000
bruit_hf = 0.2 * np.sin(2*np.pi*f_bruit*t_audio)

signal_bruite = signal_audio + bruit_hf

X_bruite = fft(signal_bruite)
amp_bruite = np.abs(X_bruite[mask])

plt.figure(figsize=(12,4))
plt.plot(freq_pos, amp_bruite)
plt.title("TP2 - Spectre du signal bruité")
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.xlim(0, 7000)
plt.grid()
plt.show()
```

```python
X_filtre = X_bruite.copy()

zone_bruit = (np.abs(freq_audio - 5000) < 50) | (np.abs(freq_audio + 5000) < 50)
X_filtre[zone_bruit] = 0

signal_filtre = np.real(ifft(X_filtre))

amp_filtre = np.abs(X_filtre[mask])

plt.figure(figsize=(12,4))
plt.plot(freq_pos, amp_bruite, label="Avant filtrage")
plt.plot(freq_pos, amp_filtre, label="Après filtrage")
plt.title("TP2 - Spectre avant/après filtrage")
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.xlim(0, 7000)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(t_audio[:2000], signal_audio[:2000], label="Original")
plt.plot(t_audio[:2000], signal_filtre[:2000], label="Filtré", alpha=0.8)
plt.title("TP2 - Comparaison temporelle")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
print("Analyse :")
print("- Le spectre du signal bruité montre un pic supplémentaire vers 5000 Hz.")
print("- Le filtrage dans le domaine fréquentiel supprime cette composante.")
print("- Le signal reconstruit est plus proche du signal original.")
```

```python
!pip -q install librosa soundfile

import librosa
from IPython.display import Audio

audio_path = librosa.ex('trumpet')
signal, fs = librosa.load(audio_path, sr=None)

print("Fréquence d'échantillonnage :", fs)
print("Nombre d'échantillons :", len(signal))

Audio(signal, rate=fs)
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(len(signal)) / fs

plt.figure(figsize=(10,4))
plt.plot(t, signal)
plt.title("Signal audio chargé")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
sig_sous = signal[::10]
fs_sous = fs // 10

print("Fréquence d'échantillonnage originale :", fs)
print("Nouvelle fréquence d'échantillonnage :", fs_sous)
print("Taille originale :", len(signal))
print("Nouvelle taille :", len(sig_sous))
```

```python
from IPython.display import Audio, display

print("Signal original :")
display(Audio(signal, rate=fs))

print("Signal sous-échantillonné :")
display(Audio(sig_sous, rate=fs_sous))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.specgram(signal, Fs=fs)
plt.title("Spectrogramme du signal original")
plt.xlabel("Temps (s)")
plt.ylabel("Fréquence (Hz)")
plt.colorbar()
plt.show()

plt.figure(figsize=(12,5))
plt.specgram(sig_sous, Fs=fs_sous)
plt.title("Spectrogramme du signal sous-échantillonné")
plt.xlabel("Temps (s)")
plt.ylabel("Fréquence (Hz)")
plt.colorbar()
plt.show()
```

```python
import numpy as np

t1 = np.arange(len(signal)) / fs
t2 = np.arange(len(sig_sous)) / fs_sous

plt.figure(figsize=(12,4))
plt.plot(t1, signal)
plt.title("Signal original")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

plt.figure(figsize=(12,4))
plt.plot(t2, sig_sous)
plt.title("Signal sous-échantillonné")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

Tu peux mettre ce texte :

Le sous-échantillonnage consiste ici à ne garder qu’un échantillon sur 10. La fréquence d’échantillonnage est donc divisée par 10. Sur le spectrogramme, on observe une dégradation du contenu fréquentiel et un repliement spectral appelé aliasing. Les sons aigus et certaines composantes fines sont altérés.

```python
!pip -q install scikit-image

from skimage import data
import matplotlib.pyplot as plt
import numpy as np

img = data.camera()   # image déjà en niveaux de gris
print(img.shape)

plt.figure(figsize=(6,6))
plt.imshow(img, cmap="gray")
plt.title("Image originale en niveaux de gris")
plt.axis("off")
plt.show()
```

```python
img_4 = (img // 64) * 85
img_2 = (img // 128) * 255

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap="gray")
plt.title("Originale (256 niveaux)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img_4, cmap="gray")
plt.title("4 niveaux de gris")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(img_2, cmap="gray")
plt.title("2 niveaux de gris")
plt.axis("off")

plt.show()
```

observation:

Quand on réduit fortement le nombre de niveaux de gris, les transitions douces disparaissent et l’image présente des bandes visibles dans les zones de dégradé. Ce défaut s’appelle banding ou contouring.

```python
from PIL import Image

img_pil = Image.fromarray(img)

w, h = img_pil.size
img_small = img_pil.resize((w//8, h//8))
img_pixel = img_small.resize((w, h), Image.NEAREST)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("Image originale")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(np.array(img_pixel), cmap="gray")
plt.title("Image pixelisée")
plt.axis("off")

plt.show()
```
