```python
N = len(signal)
fft_result = fft(signal)
freqs      = fftfreq(N, d=1/sample_rate)

# Garder seulement les fréquences positives
pos_mask  = freqs > 0
freqs_pos = freqs[pos_mask]
amplitude = np.abs(fft_result[pos_mask]) / N

# Trouver la fréquence fondamentale (pitch)
# On cherche dans la plage 80–400 Hz (voix humaine)
mask_voix = (freqs_pos >= 80) & (freqs_pos <= 400)
f0_index  = np.argmax(amplitude[mask_voix])
f0        = freqs_pos[mask_voix][f0_index]

# Afficher le spectre
plt.figure(figsize=(12, 5))
plt.plot(freqs_pos, amplitude, color='darkorange', linewidth=0.9)
plt.axvline(x=f0, color='red', linestyle='--', linewidth=1.5,
           label=f'Fréquence fondamentale f0 = {f0:.1f} Hz')
plt.xlim(0, 3000)
plt.xlabel('Fréquence (Hz)', fontsize=13)
plt.ylabel('Amplitude', fontsize=13)
plt.title('Spectre fréquentiel (FFT) — Étudiant 1', fontsize=15, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/fft_e1.png', dpi=150)
plt.show()

print(f'Fréquence fondamentale (pitch) : {f0:.1f} Hz')
if f0 < 165:
    print("   → Voix grave (type baryton/basse)")
elif f0 < 255:
    print("   → Voix medium (ténor/mezzo-soprano)")
else:
    print("   → Voix aiguë (soprano/haute)")
```

```python
n = len(signal)
seg_a = signal[n//3 * 0 : n//3 * 1]
seg_i = signal[n//3 * 1 : n//3 * 2]
seg_o = signal[n//3 * 2 : n//3 * 3]

def compute_fft(seg, sr):
    N = len(seg)
    Y = np.abs(fft(seg)) / N
    F = fftfreq(N, d=1/sr)
    mask = F > 0
    return F[mask], Y[mask]

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
voyelles = [('A', seg_a, '#e74c3c'),
            ('I', seg_i, '#2ecc71'),
            ('O', seg_o, '#3498db')]

for ax, (voy, seg, col) in zip(axes, voyelles):
    F, Y = compute_fft(seg, sample_rate)
    ax.plot(F, Y, color=col, linewidth=0.9)
    ax.set_xlim(0, 3000)
    ax.set_title('Voyelle "' + voy + '"', fontsize=14, fontweight='bold', color=col)
    ax.set_xlabel('Fréquence (Hz)')
    ax.grid(True, alpha=0.3)

axes[0].set_ylabel('Amplitude')
plt.suptitle('Comparaison spectrale des voyelles A / I / O',
             fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/voyelles_e1.png', dpi=150)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from scipy.fftpack import fft2, ifft2, fftshift
from skimage import data, img_as_float, img_as_ubyte
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

def motion_psf(shape, angle, length):
    """Crée une PSF de mouvement linéaire uniforme."""
    PSF = np.zeros(shape)
    center = (shape[0]//2, shape[1]//2)
    # Générer les points du segment
    for i in range(length):
        dx = int(round(i * np.cos(np.radians(angle))))
        dy = int(round(i * np.sin(np.radians(angle))))
        x = center[0] + dx
        y = center[1] + dy
        if 0 <= x < shape[0] and 0 <= y < shape[1]:
            PSF[x, y] = 1
    # Normalisation
    PSF /= PSF.sum()
    return PSF

def wiener_deconvolution(blurred, PSF, K=0.01):
    """Déconvolution de Wiener."""
    G = fft2(blurred)
    H = fft2(PSF)
    H_conj = np.conj(H)
    denom = np.abs(H)**2 + K
    F_est = (H_conj / denom) * G
    f_est = np.real(ifft2(F_est))
    return np.clip(f_est, 0, 255)

# Charger l'image de test
img = data.camera().astype(np.float32)  # valeurs entre 0 et 255

# Paramètres du flou
angle = 30
length = 20
PSF = motion_psf(img.shape, angle, length)

# Appliquer le flou (convolution)
from scipy.ndimage import convolve
blurred = convolve(img, PSF, mode='reflect')

# Ajouter un bruit gaussien
noise_std = 5
np.random.seed(42)
blurred_noisy = blurred + np.random.normal(0, noise_std, img.shape)

# Restauration avec différentes valeurs de K
K_values = [0.001, 0.01, 0.05, 0.1]
restored_images = []
for K in K_values:
    restored = wiener_deconvolution(blurred_noisy, PSF, K=K)
    restored_images.append(restored)

# Calcul des métriques (PSNR et SSIM) par rapport à l'originale
metrics = []
for rest in restored_images:
    psnr_val = psnr(img, rest, data_range=255)
    ssim_val = ssim(img, rest, data_range=255)
    metrics.append((psnr_val, ssim_val))

# Affichage des résultats
plt.figure(figsize=(15, 10))

plt.subplot(2,4,1)
plt.imshow(img, cmap='gray')
plt.title('Originale')
plt.axis('off')

plt.subplot(2,4,2)
plt.imshow(PSF, cmap='gray', extent=[0, img.shape[1], img.shape[0], 0])
plt.title('PSF (mouvement)')
plt.axis('off')

plt.subplot(2,4,3)
plt.imshow(blurred_noisy, cmap='gray')
plt.title(f'Floue + bruit (σ={noise_std})')
plt.axis('off')

for i, (K, rest, (p, s)) in enumerate(zip(K_values, restored_images, metrics)):
    plt.subplot(2,4,4+i)
    plt.imshow(rest, cmap='gray')
    plt.title(f'Wiener K={K}\nPSNR={p:.2f} dB, SSIM={s:.3f}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

### Resultats obtenus
| K       | PSNR     | SSIM  |
|---------|----------|-------|
| K=0.001 | 6.34 dB  | 0.012 |
| K=0.01  | 6.61 dB  | 0.045 |
| K=0.05  | 6.87 dB  | 0.142 |
| K=0.1   | 7.05 dB  | 0.196 |

### Analyse
- K petit (0.001) : image tres bruitee, PSNR faible (6.34 dB)
- K grand (0.1) : image plus stable, meilleur PSNR (7.05 dB)
- Le meilleur resultat est K=0.1 avec PSNR=7.05 dB et SSIM=0.196

### Pourquoi le filtre de Wiener ?
La division spectrale naive (1/H) amplifie le bruit
catastrophiquement. Le parametre K regularise :
H* / (|H|^2 + K)
- K petit = plus net mais amplifie le bruit
- K grand = plus stable, meilleur compromis

### Conclusion
Le filtre de Wiener permet de restaurer partiellement
une image degradee par un flou de bouge. Le PSNR
s'ameliore de 6.34 dB (K=0.001) a 7.05 dB (K=0.1),
confirmant que K=0.1 donne le meilleur resultat
pour notre image.

```python
import os
from scipy.io import wavfile
import numpy as np
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt

def compute_fft(seg, sr):
    N = len(seg)
    Y = np.abs(fft(seg)) / N
    F = fftfreq(N, d=1/sr)
    mask = F > 0
    return F[mask], Y[mask]

sr1, sig1 = wavfile.read('/content/voix_etudiant1.wav')
if sig1.ndim == 2:
    sig1 = sig1.mean(axis=1)
signal = sig1 / np.max(np.abs(sig1))
sample_rate = sr1

sr2, sig2 = wavfile.read('/content/voix_etudiant2.wav')
if sig2.ndim == 2:
    sig2 = sig2.mean(axis=1)
sig2 = sig2 / np.max(np.abs(sig2))

F1, Y1 = compute_fft(signal, sample_rate)
F2, Y2 = compute_fft(sig2, sr2)

m1 = (F1 >= 80) & (F1 <= 400)
f0_1 = F1[m1][np.argmax(Y1[m1])]
m2 = (F2 >= 80) & (F2 <= 400)
f0_2 = F2[m2][np.argmax(Y2[m2])]

plt.figure(figsize=(12, 5))
plt.plot(F1, Y1, color='steelblue', alpha=0.8, linewidth=0.9,
         label='Etudiant 1 - f0 = ' + str(round(f0_1, 1)) + ' Hz')
plt.plot(F2, Y2, color='darkorange', alpha=0.8, linewidth=0.9,
         label='Etudiant 2 - f0 = ' + str(round(f0_2, 1)) + ' Hz')
plt.axvline(f0_1, color='steelblue', linestyle='--', alpha=0.6)
plt.axvline(f0_2, color='darkorange', linestyle='--', alpha=0.6)
plt.xlim(0, 3000)
plt.xlabel('Frequence (Hz)', fontsize=13)
plt.ylabel('Amplitude', fontsize=13)
plt.title('Superposition des spectres', fontsize=15, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/superposition_spectre.png', dpi=150)
plt.show()

print("Etudiant 1 - Pitch : " + str(round(f0_1, 1)) + " Hz")
print("Etudiant 2 - Pitch : " + str(round(f0_2, 1)) + " Hz")
if f0_1 < f0_2:
    print("Voix la plus grave : Etudiant 1")
else:
    print("Voix la plus grave : Etudiant 2")
```
